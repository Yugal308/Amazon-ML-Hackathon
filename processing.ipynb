```python
import pandas as pd
import numpy as np
```

```python
reprocessed_data = pd.read_csv("title_preprocessing.csv")
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df = df.dropna()
df
```

```python
reprocessed_data['PRODUCT_LENGTH'] = df['PRODUCT_LENGTH'].values
reprocessed_data
```

```python
reprocessed_data.isnull().sum()
```

```python
reprocessed_data2 = reprocessed_data.dropna()
```

```python
reprocessed_data2.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(reprocessed_data2['text_lemmatized'], reprocessed_data2['PRODUCT_LENGTH'], test_size=0.2, random_state=30)
```

```python
tfidf = TfidfVectorizer(stop_words='english', max_df=0.8) # stop words removal and max document frequency set to 0.7
X_train = tfidf.fit_transform(X_train)
X_test = tfidf.transform(X_test)
```

```python
X_train.dtype
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestRegressor
```

```python
regr = RandomForestRegressor(max_depth=20, random_state=42)
regr.fit(X_train, y_train)
```

```python
y_pred = regr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
rsquare = r2_score(y_pred,y_test)
rsquare
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true))
```

```python
acc = 100*(1-mean_absolute_percentage_error(y_test, y_pred))
acc
```
